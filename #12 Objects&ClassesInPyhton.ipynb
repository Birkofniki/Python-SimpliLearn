# OOP- Object Oriented Programming

```python
class car():
    def getSpeed(self): # self is important as it refers to an instance, like BMW and FORD instance here
        print("155 km/h")
BMW= car()
FORD= car()
car.getSpeed(BMW) # calling the function getSpeed
car.getSpeed(FORD)

        
```

```python
# Using tyhe Keyterm _init_
class car():
    def __init__(self, year, speed):
        self.year=year
        self.speed=speed
            
    def getSpeed(self):
        print("maximum speed is:", self.speed)
        
BMW= car(2018, 155)
FORD= car(2016, 140)

BMW.getSpeed() # Does the same as above; calling the function getSpeed
FORD.getSpeed()
```

```python
# Modifying the above code to set new speed
class car():
    def __init__(self, year, speed):
        self.year=year
        self.speed=speed
            
    def getSpeed(self):
        print("maximum speed is:", self.speed)
    def setSpeed(self, speed):
        self.speed=speed
        
BMW= car(2018, 155)
FORD=car(2016, 140)

BMW.getSpeed()
BMW.setSpeed(180) #modifies or appends the speed value,, 155km/h to 180km/h
BMW.getSpeed()
FORD.getSpeed()
```

# INHERITANCE in Python programming 

```python
# Inheritance refers to the mechanism for a new class to use the features of another class
# example below
# Modifying the above code to show inheritance
class car():
    def __init__(self, year, speed):
        self.year=year
        self.speed=speed
            
    def getSpeed(self):
        print("maximum speed is:", self.speed)
    def setSpeed(self, speed):
        self.speed=speed
        
BMW= car(2018, 155)
FORD=car(2016, 140)

class sedan(car): # child class to car
    def accelerate(self):
        print("140") # sedan is faster than suv realistically
    def openTrunk(self):
        print("Trunk has been Opened")
class SUV(car): # another child class to car
    def accelerate(self):
        print("130")
Honda=sedan(2018, 150)
BMW.getSpeed()
Honda.getSpeed()


```

# ENCAPSULATION

```python
# It's the feature of preventing data from direct access
# Example is given below
class car():
    def __init__(self, year, speed):
        self.year=year
        self.speed=speed
            
    def getSpeed(self):
        print("maximum speed is:", self.speed)
    def setSpeed(self, speed):
        self.speed=speed
        # For instance all the code above is encapsulated;can't be altered............................
        # It bundles things together, such that we can just point to them or call them........................
        # Anything you do below doesn't change the class car
        
BMW= car(2018, 155)
FORD=car(2016, 140)

class sedan(car): # child class to car
    def accelerate(self):
        print("140") # sedan is faster than suv realistically
    def openTrunk(self):
        print("Trunk has been Opened")
class SUV(car): # another child class to car
    def accelerate(self):
        print("130")
Honda=sedan(2018, 150)
BMW.getSpeed()
Honda.getSpeed()
```
