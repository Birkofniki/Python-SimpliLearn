```python
import matplotlib.pyplot as plt # type: ignore
import numpy as np # type: ignore

# Create a sample data set
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y)

# Add title and labels
plt.title('Sine Wave')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Show the plot
plt.show()
```

# String Operations

```python
var="Augastino"
```

```python
#extracting a single character is same as integers, e.g
print(var[0])
```

```python
#printing the 5th character of the variable
# the index is always n-1
print(var[4])
```

```python
# to extract the first 3 characters, we do this
# the last index is always excluded
print(var[0:3])
```

```python
# can also ignore the 0
print(var[:3])
```

```python
# to print 4th character to the last one
print(var[4:])
```

```python
#print excess
print(var[0:20])
```

```python
#to know length of string
len(var)
```
