```python
# (1,2,1) and (1,2,2) are significant if we want tiled subplots, without which both the graphs will in the same plot

#For example
pylab.subplot(1,2,1)# The contents of the brackets represent (rows, columns, indexes)
pylab.plot(x,y, 'r--')# The third parameter here determines color and line style
pylab.subplot(1,2, 1)
pylab.plot(y, x, 'g*-')

# This code demonstrates how to create tiled subplots in matplotlib using the `subplot()` function from the `pylab` module. The `subplot()` function takes three arguments: the number of rows, the number of columns, and the index of the subplot within the grid of subplots.

"""
1. `pylab.subplot(1,2,1)`: This line creates the first subplot in a grid with 1 row and 2 columns.
2. `pylab.plot(x,y, 'r--')`: This line plots the `x` and `y` data in the first subplot, with a red dashed line style.
3. `pylab.subplot(1,2, 1)`: This line selects the first subplot again (same as the first line).
4. `pylab.plot(y, x, 'g--')`: This line plots `y` against `x` in the first subplot, with a green dashed line style.

Note that the second call to `pylab.subplot(1,2, 1)` is redundant, as it selects the same subplot that was already selected in the first call. This will cause the second plot to overwrite the first plot in the same subplot.
"""
```

```python
# To create two separate subplots, you should use `pylab.subplot(1,2, 2)` for the second subplot, like this:

pylab.subplot(1,2,1)
pylab.plot(x,y, 'r--')
pylab.subplot(1,2, 2)
pylab.plot(y, x, 'g--')

#This will create two separate subplots, one with a red dashed line plot of `x` vs `y`, and the other with a green dashed line plot of `y` vs `x`.
```

```python
# Drawing contour images
import matplotlib
import numpy as np
import matplotlib.cm as cm # cm means color-maps
import matplotlib.pyplot as plt # for plotting graphs

# creating the data
delta= 0.025
x= np.arange(-3.0, 3.0, delta)
y= np.arange(-2.0, 2.0, delta)

X, Y= np.meshgrid(x,y) # these are numpy commans, revisit it
Z1= np.exp(-X**2 - Y**2)
Z2= np.exp(-(X-1)**2 - (Y-1)**2) 
Z= (Z1-Z2)*2


print(X)
print("\n","\n", Y) #the empty lines separate X and Y coordinatinates
# Z represents the height of the contour
```

```python
#continuation of the above
fig, ax= plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10) # printing this on a clabel
ax.set_title("MY CONTOUR MAP") # title is always a string, don't forget the quotes
#X and Y  are coordinates
# Z represents the height of the contour
# contour maps are not common in data science
```
